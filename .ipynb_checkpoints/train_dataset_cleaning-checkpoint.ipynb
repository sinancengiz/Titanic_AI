```python
#Dependencies
import pandas as pd
```

```python
#Read in CSV
train_data = pd.read_csv('Resources/train.csv')
train_data.head()
```

```python
#Find all null values by column to figure out how to best clean the data
train_data.isnull().sum()
```

```python
#Clean embarked data by filling forward
train_data["Embarked"] = train_data["Embarked"].fillna(method='ffill')
```

```python
#Confirm embarked data has no more null values
train_data.isnull().sum()
```

```python
#Drop name and cabin because name does not affect our algorithm, and cabin had too many null values to be included
train_data.drop(labels=['Name', 'Cabin'], axis=1, inplace=True)
```

```python
#Encode Sex column so it can be used in our algorithm
cleanup_nums = {"Sex":     {"male": 0, "female": 1}}
train_data.replace(cleanup_nums, inplace=True)
train_data.head()
```

```python
#Encode Embarked column so it can be used in our algorithm
train_data['Embarked'].unique()
cleanup_nums_2 = {"Embarked":     {"S": 0, "C": 1, 'Q': 2}}
train_data.replace(cleanup_nums_2, inplace=True)
train_data.head()
```

```python
#Drop Ticket and PassengerId because they are not useful for predicting survival rate
train_data.drop(labels=['Ticket', 'PassengerId'], axis=1, inplace=True)
train_data.head()
```

```python
#Drop any remaining null values (if they exist)
train_data.dropna(inplace=True)
train_data.head()
```

```python
#Convert this dataframe into a csv representing our cleaned data
train_data.to_csv("Resources/cleaned_train.csv", index=False)
```
